```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("LagunaSecaLapTimesMore.csv")

df
```

```python
df.columns
```

```python
df2 = df[['Time', 'PS', 'Weight_KG', 'Car_type', 'Year',
       'Country_of_Origin', 'Displacement', 
       'Engine_Layout', 'Drivetrain']]

df2
```

```python
df2.columns
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

cartypedummies = pd.get_dummies(data = df2["Car_type"], prefix = "CarType")

countrydummies = pd.get_dummies(data = df2["Country_of_Origin"], prefix = "Country")

layoutdummies = pd.get_dummies(data = df2["Engine_Layout"], prefix = "Layout")

drivedummies = pd.get_dummies(data = df2["Drivetrain"], prefix = "Drive")
```

```python
numericals = df2[['Year','PS', 'Weight_KG', 'Displacement']]

numericals
```

```python
newdf = numericals.join(cartypedummies)

newdf = newdf.join(countrydummies)

newdf = newdf.join(layoutdummies)

newdf = newdf.join(drivedummies)

newdf
```

```python
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering, SpectralClustering

# Create our scaler

scaleboi = RobustScaler()

dfx = scaleboi.fit_transform(newdf)

dfx = pd.DataFrame(dfx, columns = newdf.columns)

dfx
```

```python
# For Clusters
ncluster = [n for n in range(2, 11)]

kmeansinertia = []
```

```python
for clusternumber in ncluster:
    kman = KMeans(n_clusters = clusternumber, max_iter=1000)
    targetdf = kman.fit_predict(dfx)
    kmeansinertia.append(kman.inertia_)

kmeansinertia
```

```python
plt.figure(figsize=(10,8))

plt.plot(ncluster,kmeansinertia)
plt.scatter(ncluster,kmeansinertia)
```

We will use n_clusters = 5

```python
kman5 = KMeans(n_clusters=5, max_iter=1000)

targetdf = pd.DataFrame(kman5.fit_predict(dfx), columns = ["Cluster"])

dfxk = df2.join(targetdf)

dfxk
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize = (20,24))
# Ax1: Power
ax1.scatter(x = dfxk["PS"], y = dfxk["Time"], c = dfxk["Cluster"])
ax1.set_xlabel("PS")
ax1.set_ylabel("Time")
# Ax2: Weight
ax2.scatter(x = dfxk["Weight_KG"], y = dfxk["Time"], c = dfxk["Cluster"])
ax2.set_xlabel("Weight in KG")
ax2.set_ylabel("Time")
# Ax3: PWR to Weight ratio
ax3.scatter(x = dfxk["PS"]/dfxk["Weight_KG"], y = dfxk["Time"], c = dfxk["Cluster"])
ax3.set_xlabel("P/W Ratio")
ax3.set_ylabel("Time")

fig.suptitle("K-Means clustering distrubution")
```

There does seem to be a correlation based on time, but it is still sort of weak.

```python
from sklearn.cluster import estimate_bandwidth

bandy = estimate_bandwidth(dfx, random_state = 42)

bandy
```

```python
meanshift = MeanShift(bandwidth = bandy, max_iter = 1000, cluster_all = False)

meanshiftresults = meanshift.fit_predict(dfx)

meanshiftresults = pd.DataFrame(meanshiftresults, columns = ["Cluster"])

dfms = df2.join(meanshiftresults)

dfms
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize = (20,24))
# Ax1: Power
ax1.scatter(x = dfms["PS"], y = dfms["Time"], c = dfms["Cluster"])
ax1.set_xlabel("PS")
ax1.set_ylabel("Time")
# Ax2: Weight
ax2.scatter(x = dfms["Weight_KG"], y = dfms["Time"], c = dfms["Cluster"])
ax2.set_xlabel("Weight in KG")
ax2.set_ylabel("Time")
# Ax3: PWR to Weight ratio
ax3.scatter(x = dfms["PS"]/dfms["Weight_KG"], y = dfms["Time"], c = dfms["Cluster"])
ax3.set_xlabel("P/W Ratio")
ax3.set_ylabel("Time")

fig.suptitle("MeanShift clustering distrubution")
```

I used cluster_all = False because I would like the outliers to be placed as its own separate class. It did a good job keeping the racecars separate.

```python
Agglo = AgglomerativeClustering(n_clusters = 5)

agglores = Agglo.fit_predict(dfx)

agglores = pd.DataFrame(agglores, columns = ["Cluster"])

dfg = df2.join(agglores)

dfg
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3,1,figsize = (20,24))
# Ax1: Power
ax1.scatter(x = dfg["PS"], y = dfg["Time"], c = dfg["Cluster"])
ax1.set_xlabel("PS")
ax1.set_ylabel("Time")
# Ax2: Weight
ax2.scatter(x = dfg["Weight_KG"], y = dfg["Time"], c = dfg["Cluster"])
ax2.set_xlabel("Weight in KG")
ax2.set_ylabel("Time")
# Ax3: PWR to Weight ratio
ax3.scatter(x = dfg["PS"]/dfg["Weight_KG"], y = dfg["Time"], c = dfg["Cluster"])
ax3.set_xlabel("P/W Ratio")
ax3.set_ylabel("Time")

fig.suptitle("Agglomerate clustering distrubution")
```

```python
spec = SpectralClustering(n_clusters= 7)

specres = spec.fit_predict(dfx)

specres = pd.DataFrame(specres, columns = ["Cluster"])

dfsp = df2.join(specres)

dfsp
```
