```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("LagunaSecaLapTimesMore.csv")

df
```

```python
df.columns
```

```python
df2 = df[['Time', 'PS', 'Weight_KG', 'Car_type', 'Year',
       'Country_of_Origin', 'Displacement', 
       'Engine_Layout', 'Drivetrain']]

df2
```

```python
df2.columns
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

cartypedummies = pd.get_dummies(data = df2["Car_type"], prefix = "CarType")

countrydummies = pd.get_dummies(data = df2["Country_of_Origin"], prefix = "Country")

layoutdummies = pd.get_dummies(data = df2["Engine_Layout"], prefix = "Layout")

drivedummies = pd.get_dummies(data = df2["Drivetrain"], prefix = "Drive")
```

```python
numericals = df2[['Year','Time', 'PS', 'Weight_KG', 'Displacement']]

numericals
```

```python
newdf = numericals.join(cartypedummies)

newdf = newdf.join(countrydummies)

newdf = newdf.join(layoutdummies)

newdf = newdf.join(drivedummies)

newdf
```

```python
from sklearn.cluster import KMeans, MeanShift, AgglomerativeClustering

# Create our scaler

scaleboi = RobustScaler()

dfx = scaleboi.fit_transform(newdf)

dfx = pd.DataFrame(dfx, columns = newdf.columns)

dfx
```

```python
# For Clusters
ncluster = [n for n in range(2, 11)]

kmeansinertia = []
```

```python
for clusternumber in ncluster:
    kman = KMeans(n_clusters = clusternumber, max_iter=1000)
    targetdf = kman.fit_predict(dfx)
    kmeansinertia.append(kman.inertia_)

kmeansinertia
```

```python
plt.figure(figsize=(10,8))

plt.plot(ncluster,kmeansinertia)
plt.scatter(ncluster,kmeansinertia)
```

We will use n_clusters = 6

```python
kman5 = KMeans(n_clusters=6, max_iter=1000)

targetdf = pd.DataFrame(kman5.fit_predict(dfx), columns = ["Cluster"])

dfxk = newdf.join(targetdf)

dfxk
```

```python
plt.figure(figsize = (20,8))
plt.scatter(x=[n for n in range(0, dfxk.shape[0])], y = dfxk["Time"], c = dfxk["Cluster"])
plt.xlabel("Entry")
plt.ylabel("Time")
```

```python
plt.figure(figsize = (20,8))
plt.scatter(x = dfxk["PS"], y = dfxk["Time"], c = dfxk["Cluster"])
plt.xlabel("PS")
plt.ylabel("Time")
```

```python
plt.figure(figsize = (20,8))
plt.scatter(x = dfxk["Weight_KG"], y = dfxk["Time"], c = dfxk["Cluster"])
plt.xlabel("Weight in KG")
plt.ylabel("Time")
```

```python
plt.figure(figsize = (20,8))
plt.scatter(x = dfxk["PS"]/dfxk["Weight_KG"], y = dfxk["Time"], c = dfxk["Cluster"])
plt.xlabel("P/W Ratio")
plt.ylabel("Time")
```
