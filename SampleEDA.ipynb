```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("LagunaSecaLapTimesMore.csv")

df
```

```python
df["Car_type"].value_counts()
```

```python
df["Country_of_Origin"].value_counts()
```

```python
df["Year"].value_counts()
```

```python
YearDF = df.groupby('Year')["Time"].mean().reset_index()

YearDF
```

```python
plt.plot(YearDF["Year"], YearDF["Time"])
plt.xlabel("Year")
plt.ylabel("Time in Seconds")
plt.title("Average Time series")
```

```python
df.boxplot(by = "Country_of_Origin", column= "Time", figsize= (10,8), grid = False)
```

```python
df.boxplot(by = "Car_type", column= "Time", figsize= (10,8), grid = False)
```

```python
plt.figure()
plt.scatter(data = df[df["Engine_Layout"] == "front engine"], x = "Displacement", y = "Time") # Front Engine
plt.scatter(data = df[df["Engine_Layout"] == "middle engine"], x = "Displacement", y = "Time") # Middle Engine
plt.scatter(data = df[df["Engine_Layout"] == "rear engine"], x = "Displacement", y = "Time") # Rear Engine
plt.xlabel("EngineDisplacement")
plt.ylabel("Time")
```

```python
plt.figure()
plt.scatter(data = df[df["Drivetrain"] == "front wheel drive"], x = "Displacement", y = "Time") # FWD
plt.scatter(data = df[df["Drivetrain"] == "rear wheel drive"], x = "Displacement", y = "Time") # RWD
plt.scatter(data = df[df["Drivetrain"] == "all wheel drive"], x = "Displacement", y = "Time") # AWD
plt.xlabel("EngineDisplacement")
plt.ylabel("Time")
```
