# LapMeta Laguna Seca Lap Times

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint

ourURL = "https://fastestlaps.com/tracks/laguna-seca-post-1988"

domain = "https://fastestlaps.com"
```

## BeautifulSoup

```python
item_from_url = requests.get(ourURL).text

soup = BeautifulSoup(item_from_url)

soup.title
```

Let's find the table that matters the most

```python
tables = soup.find_all("tbody")

tables[0]
```

```python
sampleitem = tables[0].find_all('tr')[0]

sampleitem

for itemtag in sampleitem.find_all("td"):
    print(itemtag.text)
    if itemtag.a:
        print(itemtag.a["href"])
```

```python
# Create a dictionary 
entrydict = {
'Name': [],
'CarLink': [],
'Time': [],
'PS': [],
'Weight_KG': []
}
```

```python
biglist = []

for item in tables[0].find_all('tr'):
    smalllist = []
    for singular in item.find_all("td"):
        #print(singular.text)
        smalllist.append(singular.text)
        if singular.a:
            smalllist.append(domain + singular.a["href"])
        else: 
            smalllist.append("")
    pprint(smalllist)
    biglist.append(smalllist)

len(biglist[0])
```

|INDEX|       TERM|
|-----|-----------|
|0    |       Rank|
|2    |      Model|
|3    |Link to Car|
|4    |     Driver|
|6    |    LS Time|
|8    |      PS/KG|

```python
from datetime import datetime

"""
This Function will read the index regarding the PS/KG and return a list with two items.
"""
def separate_values(text):
    splittext = text.split("/")
    return [splittext[0].strip(), splittext[1].strip()]

"""
This function will use datetime function to convert string into time
"""
format = "%M:%S.%f"
def turntotime(text):
    timeset = datetime.strptime(text, format).time()
    seconds = (timeset.minute * 60) + timeset.second + (timeset.microsecond / 1e6)
    return seconds
```

```python
for item in biglist:
    entrydict["Name"].append(item[2])
    entrydict["CarLink"].append(item[3])
    seconds = turntotime(item[6][0:7])
    entrydict["Time"].append(seconds)
    numberlist = separate_values(item[8])
    try:
        entrydict["PS"].append((int)(numberlist[0]))
    except ValueError as e:
        print(f"Ayo you got an error: {e}")
        entrydict["PS"].append(0)
    try:
        entrydict["Weight_KG"].append((int)(numberlist[1]))
    except ValueError as e:
        print(f"Ayo you got an error: {e}")
        entrydict["Weight_KG"].append(0)

entrydict
```

```python
RecordsDF = pd.DataFrame.from_dict(entrydict)

RecordsDF = RecordsDF[RecordsDF["PS"] > 0]

RecordsDF = RecordsDF[RecordsDF["Weight_KG"] > 0]

RecordsDF = RecordsDF[RecordsDF["CarLink"] != ""]

RecordsDF.reset_index()

RecordsDF.to_csv("LagunaSecaLapTimes.csv", index = False)

RecordsDF
```
